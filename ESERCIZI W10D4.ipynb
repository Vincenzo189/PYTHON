```python
#ESERCIZIO 6 Il dataset insurance.csv Visualizziamone le dimensioni, un'anteprima, e osserviamo i nomi di colonna; 
#• Quali sono le medie di charges rispetto a region? Ci sono differenze significative? • E rispetto a smoker? E a sex? 
#• Quali sono i descrittori statistici di bmi? Quali sono minimo, media e massimo di charges rispetto ai diversi quartili dei valori di bmi?
```

```python
insurance = pd.read_csv("dataset/beginner_datasets/insurance.csv")
insurance
```

```python
insurance.shape
```

```python
insurance.size
```

```python
insurance.dtypes
```

```python
insurance.columns
```

```python
regioni = insurance['region'].unique()
for regione in regioni:
    solo_regione = insurance[insurance['region'] == regione]     # seleziona solo le righe con quella regione
    media = solo_regione['charges'].mean()         # calcola la media della colonna charges
    print(f"Regione: {regione}, Media charges: {media}")    
```

```python
fumatori = insurance['smoker'].unique()
for fumatore in fumatori:
    stato_fumatore = insurance[insurance['smoker'] == fumatore]
    media = stato_fumatore['charges'].mean()
    print(f"Fumatore: {fumatore}, Media charges: {media}")   
```

```python
sessi = insurance['sex'].unique()
for sesso in sessi:
    tipo_sesso = insurance[insurance['sex'] == sesso]
    media = tipo_sesso['charges'].mean()
    print(f"Sesso: {sesso}, Media charges: {media}") 
```
